```python
import pandas as pd
import numpy as np # for calculating standard deviation and mean
import scipy.stats as sp # for calculating standard error
import matplotlib.pyplot as plt # for improving our visualizations# Read data
avocado = pd.read_csv("avocado.csv")
print(avocado.head())

def double_std(array):
    return np.std(array) * 2

avocado_prices = avocado.groupby("type").agg([np.mean, np.std])
avocado_prices = avocado_prices["AveragePrice"]
print(avocado_prices)

ax = avocado_prices.plot(kind = "bar", y = "mean", legend = False, title = "Average Avocado Prices")
ax.errorbar(avocado_prices.index, avocado_prices["mean"], yerr = avocado_prices["std"], 
            linewidth = 1.5, color = "black", alpha = 0.4, capsize = 4)
```

### Visualização 3-D

```python
# gerando os eixos 3D
#from mpl_toolkits import mplot3d
#
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection="3d")
plt.show()
```

```python
# matplotlib-exemplo-30
# define uma função
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
# curva 3D
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)
#X, Y = np.meshgrid(x, y) 
Z = f(x, y)
#
fig = plt.figure(figsize=(10,7))
ax =  plt.axes(projection='3d')
ax.set_xlabel('seno-coseno')
#
ax.plot3D(x, y, Z, 'red')
plt.show()
```

```python
# superficie 3D
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)
X, Y = np.meshgrid(x, y) 
Z = f(X, Y)
#
fig = plt.figure(figsize=(10,7))
ax =  plt.axes(projection='3d')
ax.set_xlabel('seno-coseno')
#
ax.plot_surface(X, Y, Z, cmap=plt.cm.jet, rstride=1, cstride=1, linewidth=0)
plt.show()
```

```python
# matplotlib-exemplo-31
# gerando os valores usando meshgrid
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)
X, Y = np.meshgrid(x, y) 
Z = f(X, Y)
plt.contour(X, Y, Z, colors='blue'); #grafico 3-D
```

```python
# matplotlib-exemplo-32
# usando contornos
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
# matplotlib-exemplo-32
# Graficando barra de cores
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
# linha 3D + scatter
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 50)
#X, Y = np.meshgrid(x, y) 
Z = f(x, y)
#
fig = plt.figure(figsize=(10,7))
ax =  plt.axes(projection='3d')
ax.set_xlabel('seno-coseno')
#
ax.plot3D(x, y, Z, 'red')
# scatter
x = x + 0.1 * np.random.randn(50)
y = y + 0.1 * np.random.randn(50)
zdata = f(x,y)
ax.scatter3D(x, y, zdata, c=zdata, cmap='Greens')
ax.set_xlabel('seno')
plt.show()
```

```python
# grafico 3D = linha 3D + scatter
fig = plt.figure(figsize=(10,7))
ax = plt.axes(projection='3d')
# gera os pontos
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
# plot 3D
ax.plot3D(xline, yline, zline, 'red')
# scatter
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')
ax.set_xlabel('seno')
plt.show()
```

### Decorando os Gráficos com Textos, Setas e Anotações

```python
# matplotlib-exemplo-44
# decorando: texto, setas e anotações
sx = np.arange(0, 2*np.pi, .01)
sy = np.sin(sx)
plt.plot(sx, sy, color = 'r')
plt.text(0.1, -0.04, '$sin(0) = 0$')
plt.text(1.5, 0.9, '$sin(90) = 1$')
plt.text(2.0, 0, '$sin(180) = 0$')
plt.text(4.5, -0.9, '$sin(270) = -1$')
plt.text(6.0, 0.0, '$sin(360) = 1$')
plt.annotate('$sin(theta)=0$', xy=(3, 0.1), xytext=(5, 0.7),arrowprops=dict(facecolor='blue', shrink=0.05))
plt.title('Insertando texto e anotações no Gráfico')
plt.xlabel('$theta$')
plt.ylabel('$sy = sin( theta)$')
plt.show()
```

### Regresão Linear Simples

```python
# ML-exemplo-01
# Exemplo de Aprendizado de máquina
#
rng = np.random.RandomState(42)
mx = 10 * rng.rand(50)
my = 2 * mx - 1 + rng.randn(50)
plt.scatter(mx, my)
plt.title('Dados para Regressão Linear');
```

```python
# ML-exemplo-02
# escolher o modelo: LinearRegression
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model
```

```python
# ML-exemplo-03
# Arrumando os dados
X = mx[:, np.newaxis]
X.shape
```

```python
# ML-exemplo-04
# Ajuste do modelo aos dados
model.fit(X, my)
```

```python
# ML-exemplo-05
# veridicando as interseções
model.coef_
```

```python
# ML-exemplo-06
# verifica as interseções
model.intercept_
```

```python
# ML-exemplo-07
# Prognostico dos rotulos para os dados desconhecidos
xfit = np.linspace(-1, 11)
Xfit = xfit[:, np.newaxis]
yfit = model.predict(Xfit)
```

```python
# ML-exemplo-08
# visualiza os dados
plt.scatter(mx, my)
plt.plot(xfit, yfit)
plt.grid();
```
